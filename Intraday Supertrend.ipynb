```python
# import CSV file
global data
data = pd.read_csv(r'data/sbin_15min.csv')
data.fillna(method='ffill', inplace=True)
data['Date']=pd.to_datetime(data['Date'])

print(str(data['Date'][0]) + " to " +  str(data['Date'][len(data)-1]))
```

```python
#generate indicator  values
def generate_indicator_values():
    global data
    data = SuperTrend(data, period, multiplier)
    return

#strategy to buy stocks
def strategy_buy(data):
    stx = 'STX_' + str(period) + '_' + str(multiplier)
    st = 'ST_' + str(period) + '_' + str(multiplier)
    
    profit = 0
    loss = 0
    trades = 0
    win_trades =0
    lose_trades=0
    
    trade_price = 0
    holding = False
    

    for i in data.index[2:]:
        if not holding:
            #Checking supertrend crossover
            if data[stx][i] == "up" and data[stx][i - 1] == "down":
                trade_price = data['Open'][i]
                trades += 1
                holding = True
                log and print("===========================================")
                log and print("Buy on => " + str(data['Date'][i]) + " " + str(data['Open'][i]))
        else:
            if data['Date'][i] >= datetime.strptime(data['Date'][i].strftime('%Y-%m-%d')+" 15:15:00","%Y-%m-%d %H:%M:%S"):
                if data['Open'][i] > trade_price:
                    profit += data['Open'][i] - trade_price
                    holding = False
                    win_trades+=1
                    log and print("Sell on => " + str(data['Date'][i]) + " " + str(data['Open'][i]))
                    log and print("Profit =>", data['Open'][i] - trade_price)
                else:
                    loss += trade_price - data['Open'][i]
                    holding = False
                    lose_trades+=1
                    log and print("Sell on => " + str(data['Date'][i]) + " " + str(data['Open'][i]))
                    log and print("Loss =>", trade_price - data['Open'][i])
                continue
                
            if data['Open'][i] >= (trade_price + trade_price * (target/100)):
                profit += data['Open'][i] - trade_price
                holding = False
                win_trades+=1
                log and print("Sell on => " + str(data['Date'][i]) + " " + str(data['Open'][i]))
                log and print("Profit =>", data['Open'][i] - trade_price)
            elif data['Low'][i] <= (trade_price - trade_price * (stop_loss/100)) :
                loss=trade_price * (stop_loss/100)
                holding = False
                lose_trades+=1
                log and print("Sell on => " + str(data['Date'][i]) + " " + str(data['Open'][i]))
                log and print("Loss =>", trade_price * (stop_loss/100))
    print("===========================================")
    print(round((profit-loss)*shares,2),profit,loss,trades,win_trades,lose_trades)


#strategy to sell stocks
def strategy_sell(data):
    stx = 'STX_' + str(period) + '_' + str(multiplier)
    st = 'ST_' + str(period) + '_' + str(multiplier)
    
    profit = 0
    loss = 0
    trades = 0
    win_trades =0
    lose_trades=0
    
    trade_price = 0
    holding = False
    

    for i in data.index[2:]:
        if not holding:
            #Checking supertrend crossover
            if data[stx][i] == "down" and data[stx][i - 1] == "up":
                trade_price = data['Open'][i]
                trades += 1
                holding = True
                log and print("===========================================")
                log and print("Sell on => " + str(data['Date'][i]) + " " + str(data['Open'][i]))
        else:
            if data['Date'][i] >= datetime.strptime(data['Date'][i].strftime('%Y-%m-%d')+" 15:15:00","%Y-%m-%d %H:%M:%S"):
                if data['Open'][i] > trade_price:
                    profit += -data['Open'][i] + trade_price
                    holding = False
                    win_trades+=1
                    log and print("Buy on => " + str(data['Date'][i]) + " " + str(data['Open'][i]))
                    log and print("Profit =>", data['Open'][i] - trade_price)
                else:
                    loss += -trade_price + data['Open'][i]
                    holding = False
                    lose_trades+=1
                    log and print("Buy on => " + str(data['Date'][i]) + " " + str(data['Open'][i]))
                    log and print("Loss =>", trade_price - data['Open'][i])
                continue
                
            if data['Open'][i] >= (trade_price + trade_price * (target/100)):
                profit += -data['Open'][i] + trade_price
                holding = False
                win_trades+=1
                log and print("Buy on => " + str(data['Date'][i]) + " " + str(data['Open'][i]))
                log and print("Profit =>", data['Open'][i] - trade_price)
            elif data['Low'][i] <= (trade_price - trade_price * (stop_loss/100)) :
                loss=trade_price * (stop_loss/100)
                holding = False
                lose_trades+=1
                log and print("Buy on => " + str(data['Date'][i]) + " " + str(data['Open'][i]))
                log and print("Loss =>", trade_price * (stop_loss/100))
    print("===========================================")
    print(round((profit-loss)*shares,2),profit,loss,trades,win_trades,lose_trades)
```
